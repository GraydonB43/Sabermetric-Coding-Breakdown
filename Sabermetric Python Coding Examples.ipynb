```python
# Pythagorean Winning Percentage Example:

# Pythagorean Winning Percentage formula: (Runs scored^2)/ ((Runs scored^2) + (Runs allowed^2))


# For the Boston Red Sox after 2023 Season (data taken from https://www.baseball-reference.com/teams/BOS/): 

RS_1 = 772 # The number of runs the Red Sox scored in 2023
RA_1 = 776 # The number of runs they allowed in 2023

# Red Sox 2023 Season results = 78-84 record or 0.481 winning percentage and missed the playoffs

# Pythagorean Winning Percentage calculation:
PWP_1 = (RS_1^2)/((RS_1^2) + (RA_1^2))
print(f' The Pythagorean Winning Percentage for the Boston Red Sox in 2023 = {PWP_1:.3f}')


# For the San Deigo Padres after 2023 Season (data taken from https://www.baseball-reference.com/teams/SDP/): 

RS_2 = 752 # Runs scored by Padres
RA_2 = 648 # Runs allowed by Padres

# San Diego Padres 2023 season results = finished season with 82-80 record and missed the playoffs

# Pythagorean Winning Percentage calculation:
PWP_2 = (RS_2^2)/((RS_2^2) + (RA_2^2))
print(f' The Pythagorean Winning Percentage for the San Deigo Padres in 2023 = {PWP_2:.3f}')


# For the Miami Marlins after 2023 Season (data taken from https://www.baseball-reference.com/teams/FLA/):

RS_3 = 666 # Runs scored by Marlins
RA_3 = 723 # Runs allowed by Marlins

# Miami Marlins 2023 season results = finished with 84-78 record and made the playoffs

# Pythagorean Winning Percentage calculation:
PWP_3 = (RS_3^2)/((RS_3^2) + (RA_3^2))
print(f' The Pythagorean Winning Percentage for the Miami Marlins in 2023 = {PWP_3:.3f}')

# Based on these Pythagorean Winning Percentages, we can make some interesting conculsions. For example, the Red
# Sox basically performed how they were expected to in 2023. Sure, their PWP tells us that they should have won 1
# or two more games, but their winning perecentage was very close to thier PWP, indicating they most likely aren't
# a playoff-caliber team. On a more interesting note, the Padres PWP indicates that they underperformed in 2023
# when they finished 82-80 by about 8 wins. This means given their run differential that they should have been a 
# playoff team in 2023. The Marlins PWP, on the other hand, indicates that they vastly overperformed in 2023 and 
# should have been an even worse team than the Red Sox. This indicates that San Diego will likely be a playoff 
# contender again after missing the playoffs this past season, whereas the Marlins may come back to earth 
# performance-wise next season.
```
